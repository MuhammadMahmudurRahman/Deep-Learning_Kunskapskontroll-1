```python
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np
```

Ladda den förtränade ResNet50-modellen med vikter tränade på ImageNet-dataset

```python
model = ResNet50(weights='imagenet')
```

```python
# img_path = 'zebra.jpg'
# img_path = 'linjal.jpg'
img_path = 'tennisboll.jpg'
img = image.load_img(img_path, target_size=(224, 224))
img
```

```python
x = image.img_to_array(img)
x.shape
```

```python
x = np.expand_dims(x, axis=0)
x.shape
```

```python
# https://www.tensorflow.org/api_docs/python/tf/keras/applications/resnet50/preprocess_input
x = preprocess_input(x)
```

```python
preds = model.predict(x)
# decode the results into a list of tuples (class, description, probability)
# (one such list for each sample in the batch)
print('Predicted:', decode_predictions(preds, top=3)[0])
# Predicted: [(u'n02504013', u'Indian_elephant', 0.82658225), (u'n01871265', u'tusker', 0.1122357), (u'n02504458', u'African_elephant', 0.061040461)]
```

# 1. Take some pictures yourself and predict their class.

### Pictures taken: Remote control, orange and mug

Describing the path to the input image:

```python
# img_path_1 = 'orange.jpg'
# img_path_1 = 'mug.jpg'
img_path_1 = 'remote.jpg'
img_1 = image.load_img(img_path_1, target_size=(224, 224))
img_1
```

Converting above image into a NumPy array:

```python
x1 = image.img_to_array(img_1)
x1.shape
```

Adding an extra dimension to the array to create a batch of size 1, since ResNet50 expects input images in batch format:

```python
x1 = np.expand_dims(x1, axis=0)
x1.shape
```

Preprocessing the image according to the requirements of the ResNet50 model:

```python
x1 = preprocess_input(x1)
```

```python
preds_1 = model.predict(x1)
```

Decodes the predicted probabilities into human-readable labels. It takes the predicted probabilities (preds) and returns a list of tuples, each containing the class ID, class description, and probability. 'Top=3': Return the top three predictions with the highest probabilities.

```python
print('Predicted:', decode_predictions(preds_1, top=3)[0])
```

# 2. Answer the question: Why is "Pretrained models" a very powerful concept?

1. Loading pretrained models that have been extensively trained on large datasets is a powerful concept because it eliminates the need to go through the entire training process, saving a lot of time.

2. Pretrained models offer the advantage of 'transfer learning', enabling users to 'fine tune' them for specific tasks or datasets. By adapting pretrained models to their own data, users can achieve high performance with minimal effort.

3. Pretrained models are readily available as 'open source libraries', making them readily available.
